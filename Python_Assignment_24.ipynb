```python
# Q1

""" Five plots that can be plotted using the Seaborn library and their uses are:

1. Scatter Plot
A scatter plot is used to visualize the relationship between two continuous variables. Each point on the plot represents an observation in the dataset, allowing for easy identification of correlations or patterns between the variables.

2. Line Plot
A line plot is ideal for visualizing trends over time or other continuous variables. It connects data points with lines, making it useful for showing changes and trends in data across intervals.

3. Bar Plot
A bar plot displays categorical data with rectangular bars, where the length of each bar represents the value of a corresponding category. It is commonly used to compare quantities across different categories, such as average values.

4. Box Plot
A box plot provides a visual summary of a datasetâ€™s distribution by displaying its quartiles, median, and potential outliers. It is useful for comparing distributions between multiple groups and identifying variations within datasets.

5. Heatmap
A heatmap visually represents data in a matrix format using colors to indicate values. It is often used to show correlations between different variables or to visualize complex datasets, making patterns easily identifiable. """
```

```python
# Q2

import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Response Over Time")
plt.show()
```

```python
# Q3

import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic, palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Box Plot of Age by Passenger Class")

plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic, palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Box Plot of Fare by Passenger Class")

plt.tight_layout()
plt.show()
```

```python
# Q4

import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))

sns.histplot(data=diamonds, x="price", hue="cut", bins=50, kde=True, palette="husl")

plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices with Cut Categories")
plt.show()
```

```python
# Q5

import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species", palette="husl", diag_kind="kde")

plt.show()
```

```python
# Q6

import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(12, 6))

sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm", linewidths=0.5)

plt.title("Heatmap of Flights Dataset")
plt.show()
```
