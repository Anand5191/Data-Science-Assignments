```python
# Q1
"""
Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in various formats. It was developed by John D. Hunter in 2003 and has become a fundamental tool for data visualization in the scientific Python ecosystem. Matplotlib allows users to create static, animated, and interactive visualizations, making it suitable for a wide range of applications from simple plots to complex visual representations of data.

Matplotlib is used for several reasons:
Versatile Plotting: It supports a wide variety of plot types including line plots, bar charts, histograms, scatter plots, and more.
Extensive Customization: Users can control every aspect of their plots, from colors and markers to labels and annotations.
Integration with NumPy and Pandas: It works seamlessly with these libraries, allowing for efficient data manipulation and visualization.
High-Quality Graphics: Matplotlib generates publication-ready plots with precise control over aesthetics.
Cross-Platform Compatibility: It can be used on Windows, macOS, and Linux without issues.

The Pyplot module of Matplotlib allows users to create various types of plots easily. Here are five common plot types that can be created using Pyplot:

Line Plot: Used to display trends over time or continuous data.
Bar Chart: Useful for comparing numerical values across different categories.
Histogram: Displays the distribution of numerical data by dividing it into bins.
Scatter Plot: Shows the relationship between two numeric variables by displaying points on a Cartesian plane.
Pie Chart: Represents proportions of a whole as slices of a pie. """
```

```python
# Q2

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, color='blue', marker='o')

plt.title("Scatter Plot of Random Data")
plt.xlabel("X-axis Values")
plt.ylabel("Y-axis Values")

plt.show()

```

```python
# Q3

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue', marker='o')
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2, color='red', marker='s')
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3, color='green', marker='^')
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple', marker='d')
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

```python
# Q4

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 5))
plt.bar(company, profit, color=['blue', 'red', 'green', 'purple'])
plt.xlabel("Company")
plt.ylabel("Profit (in million $)")
plt.title("Company Profits - Bar Plot")
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(company, profit, color=['blue', 'red', 'green', 'purple'])
plt.xlabel("Profit (in million $)")
plt.ylabel("Company")
plt.title("Company Profits - Horizontal Bar Plot")
plt.show()
```

```python
# Q5


import numpy as np
import matplotlib.pyplot as plt

np.random.seed(5)
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8, 5))
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"], patch_artist=True)

plt.title("Box Plot of Two Datasets")
plt.xlabel("Dataset")
plt.ylabel("Values")
plt.show()
```
