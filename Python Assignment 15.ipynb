```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to Home'

@app.route('/profile/')
def profile(username):
    return f'Profile of {username}'

with app.test_request_context():
    print(url_for('home'))
    print(url_for('profile', username='john'))

if __name__ == "__main__" :
	app.run(debug = True)

```
