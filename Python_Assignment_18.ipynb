```python
#Q1

from googleapiclient.discovery import build

def get_video_urls(channel_id, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)
    request = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        maxResults=5,
        order='date'
    )
    response = request.execute()

    video_urls = []
    for item in response['items']:
        if item['id']['kind'] == 'youtube#video':
            video_id = item['id']['videoId']
            video_urls.append(f"https://www.youtube.com/watch?v={video_id}")

    return video_urls


API_KEY = 'https://www.youtube.com/@PW-Foundation/videos'
CHANNEL_ID = '@PW-Foundation'
video_urls = get_video_urls(CHANNEL_ID, API_KEY)

print("First five video URLs:")
for url in video_urls:
    print(url)
```

```python
#Q2

from googleapiclient.discovery import build

def get_thumbnail_urls(channel_id, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)
    request = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        maxResults=5,
        order='date'
    )
    response = request.execute()

    thumbnail_urls = []
    for item in response['items']:
        if item['id']['kind'] == 'youtube#video':
            thumbnail_url = item['snippet']['thumbnails']['high']['url']
            thumbnail_urls.append(thumbnail_url)

    return thumbnail_urls


API_KEY = '@PW-Foundation'
CHANNEL_ID = '@PW-Foundation'
thumbnail_urls = get_thumbnail_urls(CHANNEL_ID, API_KEY)

print("Thumbnail URLs of the first five videos:")
for url in thumbnail_urls:
    print(url)
```

```python
#Q3

from googleapiclient.discovery import build

def get_video_titles(channel_id, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)
    request = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        maxResults=5,
        order='date'
    )
    response = request.execute()

    video_titles = []
    for item in response['items']:
        if item['id']['kind'] == 'youtube#video':
            video_titles.append(item['snippet']['title'])

    return video_titles


API_KEY = '@PW-Foundation'
CHANNEL_ID = '@PW-Foundation'
video_titles = get_video_titles(CHANNEL_ID, API_KEY)

print("Titles of the first five videos:")
for title in video_titles:
    print(title)
```

```python
#Q4

from googleapiclient.discovery import build

def get_video_views(channel_id, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Step 1: Fetch the first five video IDs
    search_request = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        maxResults=5,
        order='date'
    )
    search_response = search_request.execute()

    video_ids = []
    for item in search_response['items']:
        if item['id']['kind'] == 'youtube#video':
            video_ids.append(item['id']['videoId'])

    # Step 2: Fetch video statistics for these IDs
    stats_request = youtube.videos().list(
        part='statistics',
        id=','.join(video_ids)
    )
    stats_response = stats_request.execute()

    video_views = {}
    for item in stats_response['items']:
        video_id = item['id']
        view_count = item['statistics']['viewCount']
        video_views[video_id] = view_count

    return video_views


API_KEY = 'YOUR_API_KEY'
CHANNEL_ID = '@PW-Foundation'
video_views = get_video_views(CHANNEL_ID, API_KEY)

print("Number of views for the first five videos:")
for video_id, views in video_views.items():
    print(f"Video ID: {video_id}, Views: {views}")


```

```python
#Q5

from googleapiclient.discovery import build

def get_video_post_times(channel_id, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Step 1: Fetch the first five video details
    request = youtube.search().list(
        part='snippet',
        channelId=channel_id,
        maxResults=5,
        order='date'
    )
    response = request.execute()

    video_post_times = []
    for item in response['items']:
        if item['id']['kind'] == 'youtube#video':
            title = item['snippet']['title']
            published_time = item['snippet']['publishedAt']
            video_post_times.append(f"Title: {title}, Published At: {published_time}")

    return video_post_times


API_KEY = '@PW-Foundation'
CHANNEL_ID = '@PW-Foundation'
video_post_times = get_video_post_times(CHANNEL_ID, API_KEY)

print("Publishing times of the first five videos:")
for post_time in video_post_times:
    print(post_time)
```
